## What is an agent?
An agent is a system component that receives a task and decides actions.

## What is a multi-agent system?
Multiple specialized agents collaborate:
- Planner: breaks task into steps.
- Researcher: gathers evidence.
- Writer: drafts output.
- Critic: checks correctness and quality.

## Key terms
- Orchestration: controlling which agent runs and when.
- Reflection loop: retrying with feedback when quality is low.
- Tool call: using external functions/APIs (search, calculator, db).
- State: shared context passed between agents.
