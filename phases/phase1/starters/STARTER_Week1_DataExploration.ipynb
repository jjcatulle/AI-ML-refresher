## Math-to-Code Bridge Challenge (Beginner-Friendly)

This section teaches you how linear regression works under the hood before using a library.

### Why this matters
When you call sklearn LinearRegression, a lot of matrix math happens in the background.
If you understand this math, you can:
- debug shape errors quickly,
- trust the model output,
- explain your work in interviews.

### What you will build
You will solve linear regression in two ways:
1. Normal Equation (direct formula, no loop)
2. Gradient Descent (iterative optimization)

Then you will compare both with sklearn.

### Core idea
For each row of data, prediction is:

y_hat = theta_0 + theta_1*x_1 + theta_2*x_2 + ... + theta_p*x_p

Where:
- theta_0 is the bias (intercept)
- theta_1..theta_p are feature weights

### Matrix shape guide (super important)
- X: [n_samples, n_features]
- y: [n_samples]
- X_b (X with bias column): [n_samples, n_features + 1]
- theta: [n_features + 1]

If shapes do not match, matrix multiplication fails.

### Method 1: Normal Equation
Formula:

theta = (X_b^T X_b)^(-1) X_b^T y

This gives the best linear-regression coefficients directly.

### Method 2: Gradient Descent
Update rule:

theta = theta - learning_rate * gradient

With:

gradient = (1/n) * X_b^T (X_b*theta - y)

This takes many small steps to reduce error.

### Your tasks
1. Build toy data with known true coefficients.
2. Implement Normal Equation and print theta.
3. Implement Gradient Descent and print theta.
4. Fit sklearn LinearRegression and compare all three results.
5. Explain in 3 lines why all methods are close (or why they differ).

### Pass criteria
- Coefficients from manual methods are close to sklearn (difference less than 1e-2 to 1e-1 on toy data).
- Gradient Descent loss decreases over iterations.
- You can explain all tensor shapes without guessing.

````python
# Math-to-Code Bridge (super beginner version)
# Goal: learn how linear regression works internally before using sklearn.

# -----------------------------
# 0) Import modules (libraries)
# -----------------------------
# numpy: numerical math (arrays, matrix multiplication)
# sklearn.linear_model: ready-made linear regression model
import numpy as np
from sklearn.linear_model import LinearRegression

# -----------------------------
# 1) Create synthetic (fake) data
# -----------------------------
# Why fake data? Because we know the correct answer in advance.
# That helps us check if our math is right.
np.random.seed(42)  # fixed randomness so your result is reproducible
n_samples = 120
n_features = 3

# X is input features, shape = [rows, columns] = [samples, features]
X = np.random.randn(n_samples, n_features)

# True relationship used to generate y
# y = bias + X @ true_theta + noise
true_theta = np.array([1.5, -2.0, 0.7])
true_bias = 4.0
noise = np.random.randn(n_samples) * 0.1

y = true_bias + X @ true_theta + noise

print("X shape:", X.shape)  # (120, 3)
print("y shape:", y.shape)  # (120,)

# -------------------------------------------------
# 2) Add bias column to X (create design matrix X_b)
# -------------------------------------------------
# Linear model has intercept (bias). We add a column of 1s so the bias
# becomes part of theta and can be solved with matrix math.
X_b = np.c_[np.ones((X.shape[0], 1)), X]
print("X_b shape:", X_b.shape)  # (120, 4)

# -------------------------------------------------
# 3) Method A: Normal Equation (direct formula)
# -------------------------------------------------
# Formula:
# theta = (X_b^T X_b)^(-1) X_b^T y
# Read as:
# 1) transpose X_b
# 2) multiply by X_b
# 3) invert
# 4) multiply by X_b^T and y
# This gives one-shot solution (no loop).
theta_normal = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

print("\nNormal Equation theta:")
print(theta_normal)

# -------------------------------------------------
# 4) Method B: Gradient Descent (step-by-step)
# -------------------------------------------------
# Idea: start with wrong theta, then improve little by little.
# theta = theta - learning_rate * gradient

# Start from zeros
theta_gd = np.zeros(X_b.shape[1])
learning_rate = 0.05
num_iters = 2000

loss_history = []
n = X_b.shape[0]

for i in range(num_iters):
    # Predict using current theta
    y_pred = X_b @ theta_gd

    # Error = prediction - true value
    error = y_pred - y

    # Gradient of MSE loss
    gradient = (1 / n) * (X_b.T @ error)

    # Update theta
    theta_gd = theta_gd - learning_rate * gradient

    # Save loss to check learning progress
    mse = np.mean(error ** 2)
    loss_history.append(mse)

print("\nGradient Descent theta:")
print(theta_gd)
print("Initial loss:", round(loss_history[0], 6))
print("Final loss:", round(loss_history[-1], 6))

# -------------------------------------------------
# 5) Method C: sklearn LinearRegression (library)
# -------------------------------------------------
model = LinearRegression()
model.fit(X, y)

# sklearn gives intercept and coefficients separately
# We combine to match our theta format: [bias, w1, w2, w3]
theta_sklearn = np.r_[model.intercept_, model.coef_]

print("\nSklearn theta:")
print(theta_sklearn)

# -------------------------------------------------
# 6) Compare all 3 methods
# -------------------------------------------------
diff_normal_vs_sklearn = np.abs(theta_normal - theta_sklearn)
diff_gd_vs_sklearn = np.abs(theta_gd - theta_sklearn)

print("\n|Normal - Sklearn| per coefficient:")
print(diff_normal_vs_sklearn)

print("\n|GradientDescent - Sklearn| per coefficient:")
print(diff_gd_vs_sklearn)

# -------------------------------------------------
# 7) Your reflection questions
# -------------------------------------------------
# Answer these in a markdown cell below:
# Q1: Why did the gradient descent loss go down over iterations?
# Q2: Why are the three theta results very close?
# Q3: What happens if learning_rate is too large (example: 1.0)?
# Q4: What happens if learning_rate is too small (example: 0.000001)?
# Q5: Explain X, y, X_b, theta shapes in your own words.
````
