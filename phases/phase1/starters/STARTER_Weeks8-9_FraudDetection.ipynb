# Weeks 8-9: Fraud Detection Risk Scorer

## Beginner Start Here
Fraud detection is an imbalanced classification problem.

### Key terms
- Class imbalance: fraud is rare compared to non-fraud.
- Precision: among predicted fraud, how many are truly fraud.
- Recall: among true fraud, how many you catch.
- PR-AUC: preferred summary metric for imbalanced binary classification.
- Threshold: probability cutoff used to label fraud.

### Notebook workflow
1. Load and inspect data.
2. Build baseline model.
3. Evaluate with precision/recall/PR-AUC.
4. Tune threshold.
5. Build risk-scoring function and summary.
