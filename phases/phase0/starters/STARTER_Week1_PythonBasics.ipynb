## Section 2: Lists

A **list** is an ordered collection of items. You can store multiple values in one variable. Lists use square brackets `[ ]`.

```python
prices = [10.99, 24.50, 5.00, 89.00]
names  = ["Alice", "Bob", "Charlie"]
mixed  = [42, "hello", True]  # lists can mix types (but usually don't in ML)
```

**Why lists matter for ML:** Your dataset is essentially a list of rows. Feature values are stored as lists. Model predictions come back as lists.

````python
# ── Lists: Indexing, Slicing, Methods ────────────────────────────────────────

monthly_charges = [65.0, 79.99, 45.0, 89.99, 55.0, 79.99, 99.0]

# INDEXING: access one element (starts at 0, not 1!)
print(f"First element  (index 0): {monthly_charges[0]}")
print(f"Last element   (index -1): {monthly_charges[-1]}")
print(f"Third element  (index 2): {monthly_charges[2]}")

# SLICING: get a range of elements [start:stop] (stop is NOT included)
print(f"\nFirst 3: {monthly_charges[0:3]}")
print(f"Last 2:  {monthly_charges[-2:]}")

# COMMON LIST METHODS
print(f"\nLength: {len(monthly_charges)} values")
print(f"Sum:    {sum(monthly_charges):.2f}")
print(f"Min:    {min(monthly_charges)}")
print(f"Max:    {max(monthly_charges)}")
print(f"Average: {sum(monthly_charges)/len(monthly_charges):.2f}")

# Adding and removing items
monthly_charges.append(110.0)  # add to the end
print(f"\nAfter append(110.0): {monthly_charges}")

# Sorting
sorted_charges = sorted(monthly_charges)
print(f"Sorted: {sorted_charges}")
````

````python
# ── List Comprehensions (used CONSTANTLY in ML code) ─────────────────────────
# A list comprehension is a compact way to create a new list from an existing one.
# Syntax: [expression for item in list if condition]

charges = [65.0, 79.99, 45.0, 89.99, 55.0, 79.99, 99.0]

# Apply a discount to all charges
discounted = [c * 0.9 for c in charges]
print("Original charges:", charges)
print("10% discounted:  ", [round(c, 2) for c in discounted])

# Filter: keep only charges above $70
high_charges = [c for c in charges if c > 70]
print(f"\nCharges over $70: {high_charges}")

# Normalize: scale to 0-1 range (common in ML preprocessing)
min_c = min(charges)
max_c = max(charges)
normalized = [(c - min_c) / (max_c - min_c) for c in charges]
print(f"\nNormalized (0-1): {[round(n, 3) for n in normalized]}")
print("(0 = lowest charge, 1 = highest charge)")
````

## Section 6: Functions

A **function** is a named, reusable block of code. Instead of writing the same logic 10 times, you write it once as a function and call it 10 times.

```python
def function_name(parameter1, parameter2):
    # body — what the function does
    result = parameter1 + parameter2
    return result   # send value back to the caller

# Call it:
output = function_name(3, 5)  # output = 8
```

**Why functions matter for ML:** Every step in your ML pipeline is a function. Every preprocessing step. Every evaluation. The entire ML workflow is a pipeline of functions.

````python
# ── Functions ─────────────────────────────────────────────────────────────────

# --- Simple function ---
def calculate_discount(price, discount_percent):
    """Returns the discounted price."""
    discount_amount = price * (discount_percent / 100)
    return price - discount_amount

original = 79.99
discounted = calculate_discount(original, 15)
print(f"Original: ${original} → 15% off → ${discounted:.2f}")

# --- ML preprocessing function ---
def normalize_list(values):
    """Scale a list to the range [0, 1]."""
    min_val = min(values)
    max_val = max(values)
    return [(v - min_val) / (max_val - min_val) for v in values]

raw_charges = [45.0, 65.0, 79.99, 89.99, 110.0]
normalized = normalize_list(raw_charges)
print(f"\nRaw:        {raw_charges}")
print(f"Normalized: {[round(n, 3) for n in normalized]}")

# --- Function with default arguments ---
def evaluate_model(accuracy, threshold=0.85):
    """Evaluate model and return pass/fail."""
    if accuracy >= threshold:
        return f"PASS (accuracy={accuracy} ≥ threshold={threshold})"
    else:
        return f"FAIL (accuracy={accuracy} < threshold={threshold})"

print(f"\n{evaluate_model(0.91)}")
print(f"{evaluate_model(0.78)}")
print(f"{evaluate_model(0.78, threshold=0.75)}")  # custom threshold
````

````python
# ── ML Pipeline as Functions ──────────────────────────────────────────────────
# In real ML code, each step of the workflow is a function.
# This is a toy version of a real ML pipeline.

def load_data():
    """Step 2: Return raw data (would normally read a CSV)"""
    return [
        {"charges": 79.99, "tenure": 12, "churned": 0},
        {"charges": 45.00, "tenure": 24, "churned": 0},
        {"charges": 110.0, "tenure": 2, "churned": 1},
        {"charges": 89.99, "tenure": 6, "churned": 1},
        {"charges": 55.00, "tenure": 36, "churned": 0},
    ]

def extract_features(data):
    """Step 3: Pull out the input columns (X)"""
    return [[row["charges"], row["tenure"]] for row in data]

def extract_labels(data):
    """Step 3: Pull out the target column (y)"""
    return [row["churned"] for row in data]

def simple_predict(features, threshold=80):
    """Step 4/5: Dumb rule-based prediction (charges > threshold → churn=1)"""
    return [1 if f[0] > threshold else 0 for f in features]

def accuracy_score(predictions, labels):
    """Step 5: Calculate accuracy"""
    correct = sum(p == l for p, l in zip(predictions, labels))
    return correct / len(labels)

# ── Run the pipeline ──
print("Running ML pipeline...")
data = load_data()
X = extract_features(data)
y = extract_labels(data)
predictions = simple_predict(X)
acc = accuracy_score(predictions, y)

print(f"Labels:      {y}")
print(f"Predictions: {predictions}")
print(f"Accuracy:    {acc:.2f}")
print("\nThis is literally the shape of every ML pipeline!")
````

## Section 8: Reading Error Messages

Errors are not failures — they are information. Every Python error has two parts:
1. **The error type** — what kind of error it is
2. **The message** — what specifically went wrong

Common error types:
| Error | Meaning |
|-------|----------|
| `NameError` | You used a variable that doesn't exist yet |
| `TypeError` | You used the wrong type (e.g., math on a string) |
| `IndexError` | You accessed a list index that doesn't exist |
| `KeyError` | You accessed a dictionary key that doesn't exist |
| `ValueError` | Function got the right type but an invalid value |
| `ImportError` | Library not installed — `pip install <library>` |

````python
# ── Reading Errors: Safe demonstrations ───────────────────────────────────────
# Each block below shows an error and how to fix it.

# --- TypeError ---
try:
    result = "price: " + 79.99  # can't concatenate string + float
except TypeError as e:
    print(f"TypeError caught: {e}")
    print(f"Fix: use f-string → f'price: {79.99}'")

# --- IndexError ---
try:
    items = [1, 2, 3]
    _ = items[10]  # index 10 doesn't exist in a list of 3
except IndexError as e:
    print(f"\nIndexError caught: {e}")
    print(f"Fix: check len(items) = {len(items)} before accessing index 10")

# --- KeyError ---
try:
    d = {"a": 1, "b": 2}
    _ = d["c"]  # 'c' key doesn't exist
except KeyError as e:
    print(f"\nKeyError caught: key {e} not in dict")
    print(f"Fix: use d.get('c', default_value) → {d.get('c', 0)}")

print("\n✅ Error handling complete. Errors are just information — read them calmly.")
````
