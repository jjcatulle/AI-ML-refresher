## Challenge 8: Correlation Analysis

**TODO**: Understand feature relationships

1. Calculate correlation matrix for numeric columns
   - Hint: Use `.select_dtypes(include=[np.number])` then `.corr()`
2. Find which two features are most correlated
3. Create a heatmap visualization
   - Hint: `sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm')`

````python
corr_matrix=iris_df.select_dtypes(include=np.number).corr()
corr_matrix
````

````python
current_max=[];
for column in corr_matrix:
    for index in corr_matrix.index:
        if column!=index:
            if current_max == []:
                current_max = [corr_matrix.loc[index,column], index, column]
            elif abs(corr_matrix.loc[index,column]) > abs(current_max[0]):
                current_max = [corr_matrix.loc[index,column], index, column]
print(f"The highest correlation is between {current_max[1]} and {current_max[2]} with a correlation of {current_max[0]}")
````

````python
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
````

````python
color_map={};
for species in iris_df['species'].unique():
    color_map[species]=np.random.rand(3,)
colors=iris_df['species'].map(color_map)

plt.scatter(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], c=colors)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Sepal Width')

plt.show()
````
